```python
import spacy
nlp_1=spacy.load('C:/Users/Prem/ml projectt/Heuro_Task_project/models')
```

```python
import copy
documents = []
txt_id = 1950
for txt in range(50):
    txt_path = 'plan50\\0'+str(txt_id)+'.txt'
    txt_id += 1
    doc = open(txt_path,'r').readlines()
    documents.append(copy.copy(doc))    
```

```python
import json
data = [json.loads(line) for line in open('final.json', 'r')]
```

```python
import pandas as pd
```

```python
rooms_data = []
for i in data:
    rooms_data += [tuple(k.values())[3] for k in data[0]['annotation'] if tuple(k.values())[2]=='ROOM']
```

```python
uniqueRooms = set(rooms_data)
```

```python
info_table = pd.DataFrame(columns=['Room type', 'Link'])
for document in documents:
    tested_item = nlp_1(document[0])
    for ent in tested_item.ents:
        if ent.label_ == 'CONNECTION':
            labels = nlp_1(ent.text)
            connected_rooms = []
            for token in labels:
                if(str(token) in uniqueRooms):
                    connected_rooms.append(str(token))
            for i in range(len(connected_rooms)-1):
                info_table.loc[len(info_table)] = [connected_rooms[i],connected_rooms[i+1]]
```

```python
info_table.drop_duplicates(keep='first',inplace=True)
```

```python
info_table
```

```python
vertices = tuple(set(list(info_table['Room type'])+list(info_table['Link'])))
```

```python
import numpy as np
adjecency_matrix = np.zeros((len(vertices),len(vertices)))    
```

```python
for i in info_table.index:
    adjecency_matrix[vertices.index(info_table['Room type'][i]),vertices.index(info_table['Link'][i])] = 1
    adjecency_matrix[vertices.index(info_table['Link'][i]),vertices.index(info_table['Room type'][i])] = 1
```
